```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

stemmer_obj = PorterStemmer()
docs = nltk.sent_tokenize(corpus)

for var in range(len(docs)):
    words = nltk.word_tokenize(docs[var])
    words = [stemmer_obj.stem(word) for word in words if word not in set(stopwords.words('english'))]
    docs[var] = " ".join(words)

docs
```

```python
## use lemmatization
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

lemmatizer = WordNetLemmatizer()
docs = nltk.sent_tokenize(corpus)

for var in range(len(docs)):
    words = nltk.word_tokenize(docs[var])
    words = [lemmatizer.lemmatize(word,pos='v') for word in words 
             if word not in set(stopwords.words('english'))]
    docs[var] = " ".join(words)

docs
```

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

data = ['food','good','bad']

data = np.array(data).reshape(-1,1) # 2D

encoder_obj = OneHotEncoder(sparse_output=False)
one_hot = encoder_obj.fit_transform(data)
print(encoder_obj.categories_)
print('')
print(one_hot)
```
