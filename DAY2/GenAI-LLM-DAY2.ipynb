```python
# step-1
from langchain_community.document_loaders import TextLoader
loader = TextLoader("my_docs.txt")
docs = loader.load()


# step-2
from langchain_text_splitters import CharacterTextSplitter
splitted_obj = CharacterTextSplitter(chunk_size=100,chunk_overlap=20)
splitted_docs = splitted_obj.split_documents(docs)
```

```python
# step-3 - Embedding
# ---------------------

from langchain_community.embeddings import OllamaEmbeddings
embedded_obj = OllamaEmbeddings(model="nomic-embed-text:latest")
```

```python
FAISS
- VectorDB
- In memory based process

from langchain_community.vectorstores import FAISS
FAISS.from_documents(<splitted_docs>,<embedded_object>) -->stored_to_vector_db
                     |_______________________________|
```

```python
## Step - 4
from langchain_community.vectorstores import FAISS
FAISS.from_documents(splitted_docs,embedded_obj)
```

```python
vector_db = FAISS.from_documents(splitted_docs,embedded_obj)
```

```python
## Step - 5 - Test
vector_db.similarity_search("what is langchain?")
```

```python
vector_db.similarity_search("How to write hello world program in C language")
```
